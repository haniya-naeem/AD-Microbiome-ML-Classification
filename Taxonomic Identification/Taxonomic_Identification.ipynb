Sort SRR Runs Using Metadata File

```python
import pandas as pd
```

```python
# Read Metadata csv file from folder into pandas dataframe
metadata = pd.read_csv(os.path.join(tax_id, "FASTQ Metadata.csv"))
# Create a new dataframe for human samples, containing only SRR runs where "Host" column == "Homo sapiens"
metadata_human = metadata[metadata["HOST"] == "Homo sapiens"]
# Create a new dataframe for mouse samples, containing only SRR runs where "Host" column == "C57BL/6"
metadata_mouse = metadata[metadata["HOST"] == "C57BL/6"]
```

```python
# Create empty dataframes for each disease condition
# Healthy:
healthy_human_data = pd.DataFrame()
# Mild cognitive impairment(MCI)
mci_human_data = pd.DataFrame()
# Alzheimer's disease (AD)
ad_human_data = pd.DataFrame()

# Use a for loop to enumerate (get index = sample and value = condition) through each row in the "Library Name" column
# For the metadata, Library Name refers to the disease condition. It is important to note that the following code only works
# for the human samples, the mouse samples have different Library Names (however, we are not using them in this project)
for sample, condition in enumerate(metadata_human["Library Name"]):
  # Check to see if the first two letters of the Library Name are "NC", denoting a normal control (healthy) sample
  # if True, use iloc to find the sample row and concatenate it to the healthy_human_data dataframe
  # if False, move to the next elif statement
  if condition[0] == "N" and condition[1] == "C":
      healthy_human_data = pd.concat([healthy_human_data, metadata_human.iloc[[sample]]])
  # Check to see if the first two letters of the Library Name are "MC", denoting a mild cognitive impairment (MCI) sample
  # if True, use iloc to find the sample row and concatenate it to the mci_human_data dataframe
  # if False, move to the next elif statement
  elif condition[0] == "M" and condition[1] == "C":
      mci_human_data = pd.concat([mci_human_data, metadata_human.iloc[[sample]]])
  # Lastly, check to see if the first two letters of the Library Name are "AD", denoting an Alzheimer's disease (AD) sample
  # if True, use iloc to find the sample row and concatenate it to the ad_human_data dataframe
  elif condition[0] == "A" and condition[1] == "D":
      ad_human_data = pd.concat([ad_human_data, metadata_human.iloc[[sample]]])
```

```python
# Make a list of the SRR run values (Run column) for each dataframe
healthy_srr_runs = list(healthy_human_data["Run"])
mci_srr_runs = list(mci_human_data["Run"])
ad_srr_runs = list(ad_human_data["Run"])
```

Download and Split Files Using SRA Toolkit

```python
from pathlib import Path
import subprocess

# Create a parent class to handle fetching and spliting SRA files for the microbiome samples using the SRA Toolkit.
class Microbiome_Sample_Processing:
  def __init__(self, srr_runs, condition = "unknown"):
    '''
    Constructor Method: Creates instance of class with the following attributes:
          self - reference to the class object
          condition - disease condition of the samples, defaults to "unknown", type = string
          srr_runs - list of SRR run IDs of the samples, type = list of strings
    '''
    self.condition = condition
    self.srr_runs = srr_runs

    # Create a subfolder in the local Google Colab /content folder named /condition_split
    # FASTQ files will be initially downloaded in the local /content folder
    self.split_runs = Path("/content") / f"{self.condition}_split"
    os.makedirs(self.split_runs, exist_ok=True)

    # Creat a folder to store temporary files (using /tmp) during the download
    self.temp_dir = Path("/tmp/microbiome_temp")
    os.makedirs(self.temp_dir, exist_ok=True)

    # Creat a subfolder in our Taxonomic_Identification Drive folder named /condition_split
    # FASTQ files will be moved to this Drive folder
    self.drive_split_runs = Path(tax_id) / f"{self.condition}_split"
    os.makedirs(self.drive_split_runs, exist_ok=True)

  def split_paired_runs(self):
    '''
    Method that when called uses SRA Toolkit to download, split, and save the split FASTQ files
    Argument: self - reference to the class object
    '''
    # Iterates through each SRR run (SRA accession number) from the list self.srr_runs
    for srr_run in self.srr_runs:
      # Use subprocess.run(argument, check=True) to run the necessary SRA Toolkit commands
      # Commands are provided as a list argument, check=True is used to raise an error if the given
      # command returns a non-zero exit code
      subprocess.run([
          "fasterq-dump", srr_run,                       # Tool to download SRR as .sra files using accession number (srr_run) associated with sample
          "--split-files",                               # Command that separates each .sra file containing paired-end runs into _1.fastq (forward reads) and _2.fastq (reverse reads) files
          "--outdir", str(self.split_runs),              # Saves .fastq files to local /content folder
          "--temp", str(self.temp_dir),                  # Saves temporary files in /tmp/microbiome_temp folder
          "--progress"],                                 # Prints progress of download
          check=True)

      # Use subprocess.run() to copy files into the Google Drive folder
      subprocess.run([
          "cp",                                          # Command that copies the files
          str(self.split_runs / f"{srr_run}_1.fastq"),   # Copies forward read file in /content
          str(self.split_runs / f"{srr_run}_2.fastq"),   # Copies reverse read file in /content
          str(self.drive_split_runs)],                   # Pastes to Drive folder
          check=True)

      # Use subprocess.run() to remove files from local /content folder
      subprocess.run([
          "rm",                                          # Command that removes the files
          str(self.split_runs / f"{srr_run}_1.fastq"),   # Removes forward read file in /content
          str(self.split_runs / f"{srr_run}_2.fastq")],  # Removes reverse read file in /content
          check=True)
```

```python
class Healthy(Microbiome_Sample_Processing):
    def __init__(self, species, srr_runs):
      self.species = species
      super().__init__(srr_runs, "healthy")

class MCI(Microbiome_Sample_Processing):
    def __init__(self, species, srr_runs):
      self.species = species
      super().__init__(srr_runs, "mci")

class AlzheimerD(Microbiome_Sample_Processing):
    def __init__(self, species, srr_runs):
      self.species = species
      super().__init__(srr_runs, "ad")
```

```python
healthy_samples = Healthy("human", healthy_srr_runs)
mci_samples = MCI("human", mci_srr_runs)
ad_samples = AlzheimerD("human", ad_srr_runs)
```

```python
healthy_samples.split_paired_runs()
mci_samples.split_paired_runs()
ad_samples.split_paired_runs()
```
